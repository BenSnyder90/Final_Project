## Merge data sets

```python
file = "NPX_data1.xlsx"
df = pd.read_excel(file, skiprows=1)
```

```python
df_indexed = df.set_index("NPX Data")
```

```python
#Drop null value rows
df_cleaned = df_cleaned.dropna(how='all')
df_cleaned
```

```python
#Reset Index Values
df_cleaned = df_cleaned.reset_index()
#Rename column to SampleID for later merge
df_cleaned = df_cleaned.rename(columns={"NPX Data":"SampleID"})
df_cleaned
```

```python
#Drop unnecessary data rows
df_cleaned = df_cleaned.drop([0,1,3,164,165])
df_cleaned = df_cleaned.reset_index(drop=True)
```

```python
df_cleaned
```

```python
#Make new dataframe using Uniprot ID as headers
headers=df_cleaned.iloc[0]
new_df = pd.DataFrame(df_cleaned.values[1:], columns=headers)
new_df.columns
```

```python
#Rename and set the index to the Samples column
new_df = new_df.rename(columns={"Uniprot ID":"SampleID"})
new_df = new_df.set_index("SampleID")
```

```python
new_df
```

```python
#Read sample info data excel to dataframe
info_df = pd.read_excel("NPX_data1_info.xlsx")

info_df.head()
```

```python
#Merge dataframes together on SampleID
merged_df = new_df.merge(info_df, on="SampleID", how="inner")
merged_df
```

```python
merged_df = merged_df.drop("Project",axis=1)
merged_df
```

```python
#Save data to CSV
merged_df.to_csv("Resources/NPX_data1_merged.csv", index=False)
```

```python
#Make dataframes for the different timeframes of data
baseline_df = merged_df.loc[merged_df['Time']=="Baseline"]
week6_df = merged_df.loc[merged_df['Time']=="Week.6"]
week12_df = merged_df.loc[merged_df['Time']=="Week.12"]
baseline_df
```

```python
baseline_siteA_df = baseline_df.loc[baseline_df['Site']=="Site_A"]
baseline_siteB_df = baseline_df.loc[baseline_df['Site']=="Site_B"]
baseline_siteC_df = baseline_df.loc[baseline_df['Site']=="Site_C"]
baseline_siteD_df = baseline_df.loc[baseline_df['Site']=="Site_D"]
baseline_siteE_df = baseline_df.loc[baseline_df['Site']=="Site_E"]
```
